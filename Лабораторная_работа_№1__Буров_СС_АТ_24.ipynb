### 2. Создание математической модели нейронной сети согласно варианту

### 1. Первый (скрытый) слой

В этом слое каждый нейрон $j$ вычисляет взвешенную сумму входов $S_j^{(1)}$, а затем применяет к ней сигмоидную функцию активации для получения выхода $P_j$.

**Развернутая запись для каждого нейрона:**

**Первый нейрон скрытого слоя ($j=1$):**
$$ S_1^{(1)} = x_1 \cdot w_{11}^{(1)} + x_2 \cdot w_{12}^{(1)} + x_3 \cdot w_{13}^{(1)} $$
$$ P_1 = \frac{1}{1 + e^{-S_1^{(1)}}} $$

**Второй нейрон скрытого слоя ($j=2$):**
$$ S_2^{(1)} = x_1 \cdot w_{21}^{(1)} + x_2 \cdot w_{22}^{(1)} + x_3 \cdot w_{23}^{(1)} $$
$$ P_2 = \frac{1}{1 + e^{-S_2^{(1)}}} $$

**Третий нейрон скрытого слоя ($j=3$):**
$$ S_3^{(1)} = x_1 \cdot w_{31}^{(1)} + x_2 \cdot w_{32}^{(1)} + x_3 \cdot w_{33}^{(1)} $$
$$ P_3 = \frac{1}{1 + e^{-S_3^{(1)}}} $$

**Четвертый нейрон скрытого слоя ($j=4$):**
$$ S_4^{(1)} = x_1 \cdot w_{41}^{(1)} + x_2 \cdot w_{42}^{(1)} + x_3 \cdot w_{43}^{(1)} $$
$$ P_4 = \frac{1}{1 + e^{-S_4^{(1)}}} $$

---

**Обобщенная формула для первого слоя:**

Для любого нейрона $j$ ($j = 1 \dots 4$) скрытого слоя:

$$ S_j^{(1)} = \sum_{i=1}^{3} x_i \cdot w_{ji}^{(1)} $$

$$ P_j = \frac{1}{1 + e^{-S_j^{(1)}}} $$

где:
- $i$ — индекс входа ($i = 1 \dots 3$);
- $j$ — индекс нейрона в скрытом слое ($j = 1 \dots 4$);
- $w_{ji}^{(1)}$ — вес связи, идущей от $i$-го входа к $j$-му нейрону скрытого слоя.

---

### 2. Второй (выходной) слой

В этом слое каждый нейрон $q$ вычисляет взвешенную сумму выходов скрытого слоя $S_q^{(2)}$, а затем применяет линейную функцию активации (умножение на коэффициент $k$).

**Развернутая запись для каждого выхода:**

**Первый выходной нейрон ($q=1$):**
$$ S_1^{(2)} = P_1 \cdot w_{11}^{(2)} + P_2 \cdot w_{12}^{(2)} + P_3 \cdot w_{13}^{(2)} + P_4 \cdot w_{14}^{(2)} $$
$$ Y_1 = k \cdot S_1^{(2)} $$

**Второй выходной нейрон ($q=2$):**
$$ S_2^{(2)} = P_1 \cdot w_{21}^{(2)} + P_2 \cdot w_{22}^{(2)} + P_3 \cdot w_{23}^{(2)} + P_4 \cdot w_{24}^{(2)} $$
$$ Y_2 = k \cdot S_2^{(2)} $$

**Третий выходной нейрон ($q=3$):**
$$ S_3^{(2)} = P_1 \cdot w_{31}^{(2)} + P_2 \cdot w_{32}^{(2)} + P_3 \cdot w_{33}^{(2)} + P_4 \cdot w_{34}^{(2)} $$
$$ Y_3 = k \cdot S_3^{(2)} $$

---

**Обобщенная формула для второго слоя:**

Для любого выхода $q$ ($q = 1 \dots 3$) выходного слоя:

$$ S_q^{(2)} = \sum_{j=1}^{4} P_j \cdot w_{qj}^{(2)} $$

$$ Y_q = k \cdot S_q^{(2)} $$

где:
- $j$ — индекс нейрона скрытого слоя ($j = 1 \dots 4$);
- $q$ — индекс нейрона выходного слоя ($q = 1 \dots 3$);
- $w_{qj}^{(2)}$ — вес связи, идущей от $j$-го скрытого нейрона к $q$-му выходному.

---

**Итоговая формула всей сети (в общем виде):**

Выход $Y_q$ вычисляется как:

$$ Y_q = k \cdot \sum_{j=1}^{4} \left( \frac{1}{1 + e^{-\sum_{i=1}^{3} x_i \cdot w_{ji}^{(1)}}} \right) \cdot w_{qj}^{(2)} $$

где $q = 1 \dots 3$.

### 3.1. Реализация модели нейронной сети

Сначала реализуем базовую функцию `Neuron`, которая моделирует прохождение сигнала через двухслойную сеть.

**Входные параметры функции:**
* `WS`: матрица весов первого слоя (вход → скрытый слой)
* `WV`: матрица весов второго слоя (скрытый → выходной слой)
* `Xx`: массив входных сигналов во времени
* `vx`: временная ось для графиков

Внутри функции происходит последовательное вычисление состояний нейронов скрытого слоя (с сигмоидной активацией) и выходного слоя (с линейной активацией).

```python
# Импорты
import random
import numpy as np
import matplotlib.pyplot as plt

# Диапазоны весов (базовый случай)
a = -1
b = 1
```

```python
def Neuron(WS, WV, Xx, vx, weight_label=""):
    """
    Прямой проход двухслойной сети:
    Xx: (3, T)   входы (3 сигнала на T шагов времени)
    WS: (4, 3)   веса вход -> скрытый (4 скрытых нейрона)
    WV: (3, 4)   веса скрытый -> выход (3 выходных нейрона)
    weight_label: подпись диапазона весов для заголовка
    """
    Y = np.zeros(3)
    S = np.zeros(4)
    P = np.zeros(4)
    s = np.zeros(3)
    y = np.zeros((3, Xx.shape[1]))

    for t in range(0, Xx.shape[1]):
        Y = np.zeros(3)
        S = np.zeros(4)
        P = np.zeros(4)
        s = np.zeros(3)

        # Скрытый слой
        for i in range(0, 4):
            for j in range(0, 3):
                S[i] = Xx[j, t] * WS[i, j] + S[i]
            P[i] = 1 / (1 + np.exp(-S[i]))

        # Выходной слой
        for i in range(0, 3):
            for j in range(0, 4):
                s[i] = P[j] * WV[i, j] + s[i]
            Y[i] = 5 * s[i]
            y[i, t] = Y[i]

    # Визуализация выхода
    plt.figure(figsize=(10, 6))
    plt.grid(alpha=0.3)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.title(f'Выходы нейронной сети | {weight_label}', fontsize=13, fontweight='bold')
    plt.plot(vx, y[0], linewidth=2, label='Y1', marker='o', markersize=4)
    plt.plot(vx, y[1], linewidth=2, label='Y2', marker='s', markersize=4)
    plt.plot(vx, y[2], linewidth=2, label='Y3', marker='^', markersize=4)
    plt.legend(fontsize=11)
    plt.tight_layout()
```

```python
def Wes(Xx, vx, a, b):
    # [0, b] - только положительные веса
    WS = np.random.uniform(0, b, (4, 3))
    WV = np.random.uniform(0, b, (3, 4))
    Neuron(WS, WV, Xx, vx, weight_label=f"Диапазон весов: [0, {b}]")

    # [a, 0] - только отрицательные веса
    WS = np.random.uniform(a, 0, (4, 3))
    WV = np.random.uniform(a, 0, (3, 4))
    Neuron(WS, WV, Xx, vx, weight_label=f"Диапазон весов: [{a}, 0]")

    # [a, b] - смешанные веса
    WS = np.random.uniform(a, b, (4, 3))
    WV = np.random.uniform(a, b, (3, 4))
    Neuron(WS, WV, Xx, vx, weight_label=f"Диапазон весов: [{a}, {b}]")
```

### 3.2. Подготовка и визуализация входных данных

Сформируем тестовые сигналы, которые будем подавать на вход сети:
1.  **Линейный сигнал** (y = kx + b)
2.  **Пороговый сигнал** (ступенька)
3.  **Синусоидальный сигнал** (y = sin(x))

Построим графики этих входных сигналов, чтобы видеть, с чем будет работать наша нейронная сеть.

```python
k = random.uniform(-10, 10)

Xx = np.random.uniform(0, 1, (3, 20))
vx = np.linspace(-20, 20, 20)

for i in range(0, 20):
    Xx[0, i] = vx[i] * k + 3
    Xx[1, i] = 0.2 if vx[i] < 0 else 1
    Xx[2, i] = np.sin(vx[i])
```

```python
plt.figure()
plt.title('Линейная функция на вход:')
plt.plot(vx, Xx[0], 'bo-', label='y = kx+3')
plt.legend()

plt.figure()
plt.title('Пороговая функция на вход:')
plt.plot(vx, Xx[1], 'r*-', label='y = 0.2 if x < 0 else 1')
plt.legend()

plt.figure()
plt.title('Синусоидальная функция на вход:')
plt.plot(vx, Xx[2], 'r*-', label='y = sin(x)')
plt.legend()
```

### 5. Контрольные вопросы

1.  **Понятие нейрона. Его математическая модель.** Искусственный нейрон — это вычислительный элемент, который принимает входные данные, вычисляет взвешенную сумму и применяет функцию активации.
y=f(wTx+b)
2.  **Нейронная сеть.**
Нейронная сеть — это совокупность связанных между собой нейронов, предназначенная для обработки и анализа данных. Объединение нейронов позволяет решать сложные нелинейные задачи, которые невозможно решить одним нейроном.
3.  **Свойства сигмоидальной функции активации.** Сигмоида ограничена диапазоном (0,1), является гладкой и дифференцируемой. Она удобна для градиентных методов обучения, так как имеет простую производную и интерпретируется как вероятность.
4.  **Области применения НС.** Распознавание изображений, обработка речи, прогнозирование временных рядов
5.  **Однослойные и многослойные НС.** Однослойные сети решают только линейно разделимые задачи, тогда как многослойные сети способны моделировать сложные нелинейные зависимости благодаря скрытым слоям.
6.  **Линейность vs Нелинейность.** Линейные функции не позволяют моделировать сложные зависимости, так как их комбинация остаётся линейной. Нелинейные функции (например, сигмоида, ReLU) позволяют сети аппроксимировать сложные функции и решать реальные задачи классификации и регрессии.
